```python
import os
import zipfile
import matplotlib.pyplot as plt
import numpy as np

dir = 'llama_index/'  # Update this to where you want to unzip the files

# Get the file sizes for the extracted files
file_sizes = {}
for file_path in os.listdir(dir):
    full_path = os.path.join(dir, file_path)
    file_size = os.path.getsize(full_path)
    file_name = os.path.basename(full_path)
    file_sizes[file_name] = file_size

# Sort the file sizes
sorted_file_sizes = sorted(file_sizes.values())

# Generate x-axis labels
x_labels = np.arange(1, len(sorted_file_sizes) + 1)

# Plot the curve with a logarithmic y-axis
plt.figure(figsize=(8, 6))
plt.plot(x_labels, sorted_file_sizes, marker='o', linestyle='-')
plt.yscale('log')
plt.title('File Sizes in the llama_index Documentation (Log Scale)')
plt.xlabel('File Index (Sorted by Size)')
plt.ylabel('File Size (Bytes, Log Scale)')
plt.grid(True)
plt.show()
```
